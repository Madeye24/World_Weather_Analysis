```python
%matplotlib inline

# Dependencies

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
plt.show()
```

```python
# 1. Add error bars
stdev = pyber_ride_df['Avg. Fare ($USD)'].std()
stdev
```

```python
#2. Add caps to the error bars
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)", yerr=stdev, capsize=3)
plt.show()
```

```python
# 3. Change the color of the bars to sky blue
pyber_ride_df.plot.bar(x="Month", 
                       y="Avg. Fare ($USD)", 
                       yerr=stdev, 
                       capsize=3, 
                       color="#75bbfd")
plt.show()
```

```python
#4. Rotate the labels on the x-axis to horizontal
pyber_ride_df.plot.bar(x="Month", 
                       y="Avg. Fare ($USD)", 
                       yerr=stdev, 
                       capsize=3, 
                       color="#75bbfd")

plt.xticks(rotation=0)

plt.show()
```

```python
#5. Set the y-axis increment to every $5
pyber_ride_df.plot.bar(x="Month", 
                       y="Avg. Fare ($USD)", 
                       yerr=stdev, 
                       capsize=3, 
                       color="#75bbfd")

plt.xticks(rotation=0)

plt.yticks(np.arange(0, 55, step=5.0))

plt.show()
```
